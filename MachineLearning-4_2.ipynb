```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sklearn as sk
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model

df=pd.read_csv('data/sonar3.csv',header=None)
df.tail()
```

```python
X= df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
model = Sequential()
model.add(Dense(24, input_dim= X.shape[1], activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, batch_size=10)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle=True,test_size=0.3,random_state=42) 
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
model = Sequential()
model.add(Dense(10, input_dim= X_train.shape[1], activation='relu')) #ê³¼ì í•©ì„ ì¤„ì´ê¸° ìœ„í•´ Denseì˜ ìˆ˜ì™€ epochsì˜ ìˆ˜ë¡œ ê¸°ê¸° ì„±ëŠ¥ ì €í•˜ ê°€ëŠ¥
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
oooooooostory= model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
score= model.evaluate(X_test, y_test) # ì„±ëŠ¥ í™•ì¸ ëª…ë ì–´ _ë¨¸ì‹ ëŸ¬ë‹ì—ì„œ score() í•¨ìˆ˜_
print(score)
print(oooooooostory.oooooooostory['accuracy'][-1])
```

```python
from sklearn.model_selection import KFold

k = 5  # ëª‡ê²¹ìœ¼ëŸ¬ ì„¤ì •í• ì§€
kfold = KFold(n_splits=k, shuffle=True, random_state=42)  # 5ë“±ë¶„ í›„ -> 4ê°œë¥¼ train, 1ê°œë¥¼ testë¡œ ì„¤ì •í•´ì„œ í›ˆë ¨ 

acc_score =[]
for train_index, test_index in kfold.split(X):
    # print("train : ", train_index,'\n\ntest : ',test_index ,'\n\n')
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model = Sequential()
    model.add(Dense(24, input_dim= X_train.shape[1], activation='relu')) #ê³¼ì í•©ì„ ì¤„ì´ê¸° ìœ„í•´ Denseì˜ ìˆ˜ì™€ epochsì˜ ìˆ˜ë¡œ ê¸°ê¸° ì„±ëŠ¥ ì €í•˜ ê°€ëŠ¥
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    oooooooostory = model.fit(X_train, y_train, epochs=100)
    acc_score.append(model.evaluate(X_test, y_test)[1])
    

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```
