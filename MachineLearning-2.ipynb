```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x1= np.array([2,4,6,8])
x2= np.array([0,4,2,3])
y= np.array([81,93,91,97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)

plt.show()
```

```python
a1, a2, b = (0,0,0)  #    
lr=0.01
epochs=2001
n=len(x1)
# y_pred= a1*x1 + a2*x2 + b
```

```python
# 
for i in range(epochs):  #  (MSE)
    y_pred = a1*x1 + a2*x2 + b
    error = y- y_pred
    
    a1=a1-lr * ((2/n)* sum(-x1*(error))) #a1 
    a2=a2-lr * ((2/n)* sum(-x2*(error))) #a2 
    b=b - lr * ((2/n)* sum(-(error)))    #b 
    
    if i % 100 == 0:
        print('epochs : {}, a1 : {:3.2f}, a2 : {:3.2f}, b : {:3.2f}'.format(i,a1,a2,b))
```

```python
#   
y_pred =a1*x1 + a2*x2 +b
print(' ',y)
print(' ',y_pred)

#  5,   3   
print('  : ',a1*5 + a2*3 + b)
```

```python
#   
model = Sequential()
model.add(Dense(1,input_dim=1, activation='linear'))     # layer 
model.summary()  # layer bais  
```

```python
# XOR  Ver
w11 = np.array([-2,-2]) # X1 
w12 = np.array([2,2])   # X2 
w2 = np.array([1,1]) # B1 
b1 = 3
b2 = -1
b3 = -1
```

```python
#   
def MLP(x,w,b):
    y = np.sum(w*x)+b
    if y <=0:
        return 0
    else:
        return 1

# NAND (?)
def NAND(x1,x2):
    return MLP(np.array([x1,x2]),w11,b1)

# OR (?)
def OR_(x1,x2):
    return MLP(np.array([x1,x2]),w12,b2)

# AND (?)
def AND(x1,x2):
    return MLP(np.array([x1,x2]),w2,b3)

# XOR 
def XOR(x1,x2):
    return AND(NAND(x1,x2),OR_(x1,x2))
```
