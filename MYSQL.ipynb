```python
import mysql.connector

mydb = mysql.connector.connect(host='localhost', user='root', password='', database='shopdb')

mycursor = mydb.cursor()
```

```python
mycursor.execute("Select * from membertb")

rows = mycursor.fetchall()
for row in rows:
    print(row)
```

```python
mydb.close()
```

```python
sql = """CREATE TABLE membertbl (
  memberID char(8) NOT NULL,
  memberName char(5) NOT NULL,
  memberAddress char(20) DEFAULT NULL,
  PRIMARY KEY (memberID)
) """

mycursor.execute(sql)   # 위 명령어 실행 (table 생성)
```
